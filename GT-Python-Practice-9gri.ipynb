# 1 Import Packages

```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt

#Create air and gas mixture object
air=ct.Solution('gri30.yaml')

# Get all of the Species objects defined in the San Diego mechanism
species = {S.name: S for S in ct.Species.list_from_file("gri30.yaml")}

# Create an ideal gas phase object with species representing complete combustion
complete_species = [species[S] for S in ("CH4", "H2", "O2", "N2", "CO2", "H2O")]
fuel1 = ct.Solution(thermo="ideal-gas", species=complete_species)

# Creates Air-Fuel Mixture for Stoichiometric Air-Fuel Calculation
fuel_stoic=ct.Solution('gri30.yaml')
```

# 2 Stage 1 To Stage 6 Turbine

```python
# Define Inlet Conditions to Compressor
P1 = 100000  # Pa (standard pressure)
T1 = 298  # K (standard temperature)

# Compressor Efficiency
Isen_Eff_Comp = 0.90
CH4_Molecular_Weight=12.011+(4*(1.00784)) #Methane Molecular Weight
H2_Molecular_Weight=2*(1.00784) #Hydrogen Molecular Weight
Air_Molecular_Weight=(2*(15.999))+(2*(14.0067)) # O2 and N2 Air Molecular Weight
fuel_mole_fraction_stoic={'CH4': 1, 'H2': 1}
air_mole_fractions_stoic = {'O2': 0.525, 'N2': 1.975}
fuel_stoic.set_mixture_fraction(1, fuel_mole_fraction_stoic, air_mole_fractions_stoic)
air_fuel_stoic_ratio=fuel_stoic.stoich_air_fuel_ratio(fuel=fuel_mole_fraction_stoic, oxidizer=air_mole_fractions_stoic)

#For Plots
equi_phi_vals = []
T3_vals = []
Cycle_Eff_vals = []
H2_Frac_vals = []
Pressure_Ratio_vals = []
equi_phi_vals = []
T3_vals = []
Cycle_Eff_vals = []
H2_Frac_vals = []
Pressure_Ratio_vals = []

#Compressor Physical Parameters
Compressor_Volume=200 # [m^3] From Google Gemini AI Search Results For Compressor Volume of 110MW
Comp_Inlet_Air_Mass_Flow = 401.7 # [kg/s]

# Loop over Compressor Pressure Ratio (10 to 20)
for PR in range (10,21,1):  # Pressure ratio from 10 to 20 in steps of 1

#Combustion in Combustion Chamber
    # Loop over Equivalence Ratio from 0 to 1 in steps of 0.1
    for equi_phi in np.arange (0.1,1.1,0.1):
        
        # Loop over Hydrogen Percentage (0% to 100%)
        for H2_frac in np.arange (0,1.1,0.1):  # Hydrogen fraction from 0% to 100% in steps of 10%
            CH4_frac = 1 - H2_frac  # Adjust methane fraction
            
            # Define fuel mole fractions
            fuel_mole_fraction={'CH4': CH4_frac, 'H2': H2_frac}
            
            # Set fuel-air mixture
            air_mole_fractions = {'O2': 0.21, 'N2': 0.79}

            # Isentropic Compression from P1 to P2
            # Compressor Work With Air Flow Accounted For. In Compressor, Gas Is Only Air
            # Set Compressor Inlet State
            air.TPX=T1,P1, air_mole_fractions

            H1=air.h      
            P2=P1*PR
            S1 = air.entropy_mass
            S2 = S1
            air.SP=S2,P2
            T2=air.T
            H2=air.h

            #Simulate the Combustion to Calculate Temperature, T3 of The Combustion Gas
            fuel1.set_equivalence_ratio(phi=equi_phi, fuel=fuel_mole_fraction, oxidizer=air_mole_fractions)

            #Set Mixture Temperature and Pressure To Premix Air and Fuel
            fuel1.TP=T2,P2

            #Find Mole Fractions of Reactants
            mole_fraction_CH4 = fuel1['CH4'].X[0]
            mole_fraction_H2 = fuel1['H2'].X[0]

            mass_of_CH4 = mole_fraction_CH4 * CH4_Molecular_Weight

            mass_of_H2 = mole_fraction_H2 * H2_Molecular_Weight

            mass_fuel = mass_of_CH4 + mass_of_H2
        
            # Compressor Air
            Comp_Mass_Air=mass_fuel/(equi_phi*air_fuel_stoic_ratio) # [kg]
            
            mass_cool=0
            mass_air_total=Comp_Mass_Air + mass_cool # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature        
            
            # Compressor Work
            W_Comp=mass_air_total*Isen_Eff_Comp*(H2-H1)

            fuel1.equilibrate('HP')

            # Back To Fuel Combustion After Getting Air Parameters
            T3 = fuel1.T
            S3=fuel1.entropy_mass            
            fuel1.TP = T3, P2
            H3=fuel1.h

            T3_Max = 2000 #Maximum Turbine Inlet Temperature           

            # Determining Whether Combustion Gas Temperatures are Above Maximum Turbine Inlet Temperature
            if T3<=T3_Max:
                # Recalculate Compressor Air
                Comp_Mass_Air=mass_fuel/(equi_phi*air_fuel_stoic_ratio) # [kg]
                
                mass_cool=0
                
                mass_fuel_total= Comp_Mass_Air + mass_cool + mass_fuel #mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

                #Compute Heat Added 
                Heat_in=mass_fuel_total*(H3-H2)

                #Turbine Section
                Isen_Eff_Turb = 0.90 # Turbine Efficiency
                S4=S3
                fuel1.SP= S4, P1
                H4=fuel1.h

                # Turbine Work
                W_Turb=mass_fuel_total*Isen_Eff_Turb*(H4-H3)
                
            else:
                mass_cool=((2e13)*T3_Max**(-3.488))*Comp_Mass_Air
                
                mass_fuel_total= Comp_Mass_Air + mass_cool + mass_fuel #mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

                #Turbine Section
                Isen_Eff_Turb = 0.90 # Turbine Efficiency
                T3=T3_Max
                fuel1.TP = T3, P2
                S3=fuel1.entropy_mass 
                H3=fuel1.h
                S4=S3
                fuel1.SP= S4, P1
                H4=fuel1.h

                #Compute Heat Added 
                Heat_in=mass_fuel_total*(H3-H2)

                # Turbine Work
                W_Turb=mass_fuel_total*Isen_Eff_Turb*(H4-H3)
            
                #Recalculate Compressor Air and the Compressor Work
                mass_air_total=Comp_Mass_Air + mass_cool # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature
                
                W_Comp=mass_air_total*Isen_Eff_Comp*(H2-H1)    
                
            #Gas Turbine Efficiency
            Net_Work = W_Turb-W_Comp
            Cycle_Eff = (Net_Work)/Heat_in

            equi_phi_vals.append(equi_phi)
            T3_vals.append(T3)
            Cycle_Eff_vals.append(Cycle_Eff)
            H2_Frac_vals.append(H2_frac)
            Pressure_Ratio_vals.append(PR)

            print('Compressor Work', W_Comp)
            print(f"Heat In: {Heat_in:.10f}")
            print(f"Turbine Work: {W_Turb:.10f}")
            print(f"Net Work: {Net_Work:.10f}")
            print(f"Cycle_Eff: {Cycle_Eff:.10f}")
    
    plt.plot(H2_Frac_vals, Cycle_Eff_vals, marker='o', linestyle='-')
    plt.xlabel('Hydrogen Fraction')
    plt.ylabel('Brayton Cycle Efficiency')
    plt.title('Cycle Efficiency With Hydrogen Fraction')
    plt.grid()
    plt.show()
```
