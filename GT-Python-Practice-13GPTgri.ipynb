# 2 Stage 1 To Stage 6 Turbine

```python
# Define Inlet Conditions to Compressor
P1 = 100000  # Pa (standard pressure)
T1 = 298  # K (standard temperature)

# Compressor Efficiency
Isen_Eff_Comp = 0.90

# Atomic Properties and Mole Fractions and Stoichiometric Combustion Parameters
CH4_Molecular_Weight=12.011+(4*(1.00784)) # Methane Molecular Weight
H2_Molecular_Weight=2*(1.00784) # Hydrogen Molecular Weight
O2_Molecular_Weight=(2*(15.999)) # O2 Air Molecular Weight
N2_Molecular_Weight=(2*(14.0067)) # N2 Air Molecular Weight
fuel_mole_fraction_stoic={'CH4': 1, 'H2': 1} #tbd
air_mole_fractions_stoic = {'O2': 0.21, 'N2': 0.79} #tbd
CH4_fuel_stoic_mole=1.0
H2_fuel_stoic_mole=1.0
O2_air_stoic_CH4_mole=2.0
O2_air_stoic_H2_mole=0.5
O2_air_vol_mole_fraction=0.21
N2_air_vol_mole_fraction=0.79

# Calculate Mass of Fuel and Air for Stoichiometric Air-Fuel Ratio. Separate Stoichiometric Reactions CH4 + 2O2 -> CO2 + 2H2O and H2 + 0.5O2 -> H2O
mole_CH4_only_stoic_reaction=1.0
mole_H2_only_stoic_reaction=1.0
CH4_stoic_mass = mole_CH4_only_stoic_reaction * CH4_Molecular_Weight
H2_stoic_mass = mole_H2_only_stoic_reaction * H2_Molecular_Weight

O2_in_air_mole_stoic_CH4_H2_separately_combined = CH4_fuel_stoic_mole * O2_air_stoic_CH4_mole + H2_fuel_stoic_mole * O2_air_stoic_H2_mole
air_total_mole_per_fuel = O2_in_air_mole_stoic_CH4_H2_separately_combined / O2_air_vol_mole_fraction
mass_air_stoic_CH4_H2_separately_combined = air_total_mole_per_fuel * (O2_air_vol_mole_fraction * O2_Molecular_Weight + N2_air_vol_mole_fraction * N2_Molecular_Weight)

air_CH4_H2_stoic_ratio = mass_air_stoic_CH4_H2_separately_combined / (CH4_stoic_mass + H2_stoic_mass)

# For Plots
equi_phi_vals = []
T3_vals = []
Cycle_Eff_vals = []
H2_Frac_vals = []
Pressure_Ratio_vals = []

# COMPRESSOR PHYSICAL PARAMETERS - tdb NOT USED REMOVE LATER
Compressor_Volume = 200 # [m^3] tbd not used in model From Google Gemini AI Search Results For Compressor Volume of 110MW 
Comp_Inlet_Air_Mass_Flow = 401.7 # tbd not used in model [kg/s]

# Loop over Compressor Pressure Ratio (10 to 20)
for PR in range (10,21,1):  # Pressure ratio from 10 to 20 in steps of 1

# Combustion in Combustion Chamber
    # Loop over Equivalence Ratio from 0 to 1 in steps of 0.1
    for equi_phi in np.arange (0.1,1.1,0.1):
        
        # Loop over Hydrogen Percentage (0% to 100%)
        for H2_frac in np.arange (0,1.1,0.1):  # Hydrogen fraction from 0% to 100% in steps of 10%
            CH4_frac = 1 - H2_frac  # Adjust methane fraction

            # Define fuel mole fractions
            fuel_mole_fraction = {'CH4': CH4_frac, 'H2': H2_frac}
            
            # Set fuel-air mixture
            air_mole_fractions = {'O2': 0.21, 'N2': 0.79}

            # Set Compressor Inlet State
            air.TPX=T1,P1, air_mole_fractions

            H1 = air.h
            S1 = air.entropy_mass
            
            # Isentropic Compression from P1 to P2
            P2 = P1*PR
            S2 = S1
            air.SP = S2, P2
            T2 = air.T
            
            # Isentropic Enthalpy After Compression
            H2s = air.h

            # Actual Enthalpy After Compression
            H2 = H1 + ((H2s-H1)/Isen_Eff_Comp)

            # Set State of Air With Actual Enthaly H2
            air.HP = H2, P2
            T2 = air.T
       
            # Find Mass of Fuel Using Mole Fraction From Actual Specified CH4_frac-H2_frac Volume Fraction Loop
            mass_of_CH4 = CH4_frac * CH4_Molecular_Weight

            mass_of_H2 = H2_frac * H2_Molecular_Weight

            mass_fuel = mass_of_CH4 + mass_of_H2            
  
            # Mass of Air From Compressor Cooling Air 
            Comp_Mass_Air = mass_fuel/(equi_phi*air_CH4_H2_stoic_ratio) # Mass of Air Compressed by Compressor for Combustion [kg]

            # If Combustor Exhaust Temperature is Lower Than The Maximum Turbine Inlet Temperature No Turbine Cooling
            mass_cool = 0

            # Mass of Combustion Gases Before Cooling, Actual Air for Combustion plus Mass of Fuel
            mass_fuel_air_in_turbine_before_cooling = Comp_Mass_Air + mass_cool + mass_fuel # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

            # Create New Air Fuel Mixture for Combustion Reaction
            fuel1 = ct.Solution('gri30.yaml')           

            # Simulate the Combustion to Calculate Temperature, T3 of The Combustion Gas
            fuel1.set_equivalence_ratio(phi=equi_phi, fuel=fuel_mole_fraction, oxidizer=air_mole_fractions)

            # Set Mixture Temperature and Pressure To Premix Air and Fuel
            fuel1.TP=T2, P2

            # Combust the Air-Fuel Mixture
            fuel1.equilibrate('TP')
            T3 = fuel1.T
            S3 = fuel1.entropy_mass
            fuel1.TP = T3, P2
            H3 = fuel1.h

            # Calculate Cooling Air if T3 Exceeds T3_Max
            T3_Max = 2000 # Maximum Turbine Inlet Temperature
            mass_cool = 0 # If Combustor Exhaust Temperature is Higher Than The Maximum Turbine Inlet Temperature

            # Determining Whether Combustion Gas Temperatures are Above Maximum Turbine Inlet Temperature
            if T3<=T3_Max:
                # Calculate Compressor Air
                Comp_Mass_Air = mass_fuel/(equi_phi*air_CH4_H2_stoic_ratio) # Mass of Air Compressed by Compressor for Combustion [kg]
                
                mass_cool = 0 # If Combustor Exhaust Temperature is Higher Than The Maximum Turbine Inlet Temperature

                mass_air_total = Comp_Mass_Air + mass_cool # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature               
                mass_fuel_air_in_turbine = Comp_Mass_Air + mass_cool + mass_fuel # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

                # Compressor Work With Air Flow Accounted For. In Compressor, Gas Is Only Air
                W_Comp=mass_air_total*Isen_Eff_Comp*(H2-H1)

                # Compute Heat Added
                Heat_in=mass_fuel_air_in_turbine*(H3-H2)

                # Turbine Section Isentropic Expansion
                Isen_Eff_Turb = 0.90 # Turbine Efficiency
                S4 = S3
                fuel1.SP = S4, P1
                
                # Isentropic Enthalpy After Expansion in Turbine                
                H4s = fuel1.h

                # Actual Enthalpy After Expansion in Turbine
                H4 = H3 - Isen_Eff_Turb * (H3 - H4s)            

                # Turbine Work
                W_Turb = mass_fuel_air_in_turbine*Isen_Eff_Turb*(H3-H4)

                # Gas Turbine Efficiency
                Net_Work = abs(W_Turb-W_Comp)
                Cycle_Eff = Net_Work/Heat_in
                
            else:
                #Start Over To Calculate Cooling Air Mass with Original Combustiong With Combustion Temperture Above Turbine Inlet Temperature as Reference and This Loop for Lower Energy at Setting Combustion Temperature to Maximumm Inlet Temperature
                # Create Air Mixture Object Again
                air = ct.Solution('gri30.yaml')

                # Define Inlet Conditions to Compressor
                P1 = 100000  # Pa (standard pressure)
                T1 = 298  # K (standard temperature)                

                # Set fuel-air mixture
                air_mole_fractions = {'O2': 0.21, 'N2': 0.79}

                # Set Compressor Inlet State
                air.TPX=T1,P1, air_mole_fractions

                H1 = air.h
                S1 = air.entropy_mass

                # Isentropic Compression from P1 to P2
                P2 = P1*PR
                S2 = S1
                air.SP = S2, P2
                T2 = air.T              
                Cooling_Air_Enthalpy = air.h 

                # Calculate Compressor Air
                Comp_Mass_Air = mass_fuel/(equi_phi*air_CH4_H2_stoic_ratio) # Mass of Air Compressed by Compressor for Combustion [kg]

                # Use Energy Balance To Calculate Cooling Air/Energy to Cool to T3 to T3_Max With Compressor Cooling Air is (H3-H3_Max) = Energy Gain by Cooling Air from T2 to T3_Max (H3_Max-H2)
                # Combust the Air-Fuel Mixture Again and Reduce Combustion Temperature to Maximum Turbine Inlet Temperature
                fuel1 = ct.Solution('gri30.yaml')
                fuel1.set_equivalence_ratio(phi=equi_phi, fuel=fuel_mole_fraction, oxidizer=air_mole_fractions)
                fuel1.TP=T3_Max, P2
                fuel1.equilibrate('TP')
                H3_Max = fuel1.h # Target Enthalpy

                # Solve for Cooling Air Mass Energy To Heat From T3_Max to Temperature Above T3 Is Equal To Energy Gained By Cooling Air To Cool The Combustion Gas from Temperature Above T3_Max to T3
                H_Above_T3_Max = H3
                mass_cool = (mass_fuel_air_in_turbine_before_cooling * (H_Above_T3_Max - H3_Max)) / (H3_Max - Cooling_Air_Enthalpy)

                mass_fuel_air_cooling_air_in_turbine = mass_fuel_air_in_turbine_before_cooling + mass_cool # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

                # Recalculate H3 by Setting it to H3_Max
                H3 = H3_Max
                T3 = T3_Max
               
                # Calculate Compressor Air and the Compressor Work
                mass_air_total = Comp_Mass_Air + mass_cool # mass_cool is Volume of Cooling Air Required if Turbine Inlet is Higher than Maximum Turbine Inlet Temperature

                W_Comp = mass_air_total*Isen_Eff_Comp*(H2-H1)
                
                # Turbine Section Isentropic Expansion
                Isen_Eff_Turb = 0.90 # Turbine Efficiency                
                S4 = S3
                fuel1.SP = S4, P1

                # Isentropic Enthalpy After Expansion in Turbine                
                H4s = fuel1.h

                # Actual Enthalpy After Expansion in Turbine
                H4 = H3 - Isen_Eff_Turb * (H3 - H4s) 

                # ReCompute Heat Added.  
                Heat_in = mass_fuel_air_cooling_air_in_turbine*(H3-H2)

                # Turbine Work
                W_Turb = mass_fuel_air_cooling_air_in_turbine*Isen_Eff_Turb*(H3-H4)

                # Gas Turbine Efficiency
                Net_Work = W_Turb-W_Comp
                Cycle_Eff = Net_Work/Heat_in

            equi_phi_vals.append(equi_phi)
            T3_vals.append(T3)
            Cycle_Eff_vals.append(Cycle_Eff)
            H2_Frac_vals.append(H2_frac)
            Pressure_Ratio_vals.append(PR)

            print('Compressor Work', W_Comp)
            print(f"Heat In: {Heat_in:.10f}")
            print(f"Turbine Work: {W_Turb:.10f}")
            print(f"Net Work: {Net_Work:.10f}")
            print(f"Cycle_Eff: {Cycle_Eff:.10f}")
    
    plt.plot(H2_Frac_vals, Cycle_Eff_vals, marker='o', linestyle='-')
    plt.xlabel('Hydrogen Fraction')
    plt.ylabel('Brayton Cycle Efficiency')
    plt.title('Cycle Efficiency With Hydrogen Fraction')
    plt.grid()
    plt.show()
```
